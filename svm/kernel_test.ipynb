```python
X, _, y = sample_from_logreg(n=250, p=20, seed=120)
n = X.shape[0]
p = X.shape[1]
```

```python
length_scale = np.linspace(1e-5, 5, 15)
inv_mul = []
for l in length_scale:
    gram = RBF(l)(X, X)
    inv_mul.append(np.linalg.norm(gram) * np.linalg.norm(np.linalg.inv(gram)))
```

```python
plt.plot(length_scale, inv_mul)
```
