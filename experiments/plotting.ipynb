```python
from og_sgd import run_sim
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_iter = 3000
err_approx = run_sim(250, 20, n_iter=n_iter)
print(err_approx.keys())
```

```python
plt.plot(np.log(np.arange(1, n_iter + 1)), err_approx['IACV'], c='black', label='IACV')
plt.plot(np.log(np.arange(1, n_iter + 1)), err_approx['NS'], c='red', label='NS')
plt.plot(np.log(np.arange(1, n_iter + 1)), err_approx['IJ'], c='blue', label='IJ')
plt.plot(np.log(np.arange(1, n_iter + 1)), err_approx['hat'], c='green', linestyle='dashed', label='CV')
plt.legend()
```
